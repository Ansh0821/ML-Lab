```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

print(df.head())
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df.drop('target', axis=1))
```

```python
kmeans2 = KMeans(n_clusters=2, random_state=42)
df['cluster_k2'] = kmeans2.fit_predict(df_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_scaled[:, 0], y=df_scaled[:, 1], hue=df['cluster_k2'], palette='viridis')
plt.title('K-Means Clustering with K=2')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

```python
kmeans3 = KMeans(n_clusters=3, random_state=42)
df['cluster_k3'] = kmeans3.fit_predict(df_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_scaled[:, 0], y=df_scaled[:, 1], hue=df['cluster_k3'], palette='viridis')
plt.title('K-Means Clustering with K=3')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

```python
kmeans4 = KMeans(n_clusters=4, random_state=42)
df['cluster_k4'] = kmeans4.fit_predict(df_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_scaled[:, 0], y=df_scaled[:, 1], hue=df['cluster_k4'], palette='viridis')
plt.title('K-Means Clustering with K=4')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```
